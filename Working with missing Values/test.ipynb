```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', np.nan, 'David'],
    'Age': [25, np.nan, 30, 22],
    'City': ['New York', 'Los Angeles', 'Chicago', np.nan]
}

df = pd.DataFrame(data)
print(df)
print("-----------------------------------------------")

# Detecting Missing Values

# Check where values are missing
print(df.isnull())


# Count total missing values per column
print(df.isnull().sum())


print("-----------------------------------------------")

#  Removing Missing Values

# Drop rows with any missing value
df_dropna = df.dropna()
print(df_dropna)
print("-----------------------------------------------")
# Drop columns with any missing value
df_dropna_col = df.dropna(axis=1)
print(df_dropna_col)

print("-----------------------------------------------")
#  Filling Missing Values

# Fill with a specific value
df_fill = df.fillna("Unknown")
print(df_fill)

# Fill numeric columns with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)

# Fill forward (propagate last valid value forward)
df_ffill = df.fillna(method='ffill')
print(df_ffill)

# Fill backward (propagate next valid value backward)
df_bfill = df.fillna(method='bfill')
print(df_bfill)

```
